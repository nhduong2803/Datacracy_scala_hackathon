```python
import pandas as pd
```

```python
retail = pd.read_csv('retail_data.csv',sep=',')
```

```python
retail.head(3)
```

### Để tìm ra dòng na
Có thể thay bằng cách tìm stockcode tương tự và điền tên vào
- retail[retail.isna().any(axis=1)]

```python
retail.describe(include='object')
```

```python
retail.info()
```

```python
# Bỏ toàn bộ dòng có ít nhất 1 giá trị na

retail = retail.dropna()
```

```python
retail.info()
```

```python
# bỏ các dòng cancel, có quantity <0
# Này cần EDA trước để tìm ra được cancel => quantity <0
retail_actual = retail[retail.Quantity >=0]
retail_actual.info()
```

```python
# lọc ra những txn đến từ UK
retail_actual_uk = retail_actual[retail_actual.Country == 'United Kingdom']
retail_actual_uk.head(3)
```

```python
# tạo thành bảng basket data gồm: transaction id, số lượng item
# unstack(level = 0 hay  -1) giúp biến đổi 2 index hoặc 2 column thành 1 column 1 index
basket_data = retail_actual_uk.groupby(['Invoice','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('Invoice')
basket_data
```

```python
# Chỉ quan tâm đến có được mua hay không, encode data => có mua = 1
def encode(x):
    if x>0:
        return 1
    else:
        return 0
basket_data = basket_data.applymap(encode)
basket_data.head()
```

```python
basket_filter = basket_data[(basket_data >0).sum(axis=1) >=2]
basket_filter
```

```python
# pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
# support là số lần item đó được mua từ bộ data nhập vào
# đang để là những item hay được mua (với độ support là từ 0.03 trở lên)
frequent_items = apriori(basket_filter, min_support = 0.03,
                        use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)
frequent_items['length'] = frequent_items['itemsets'].apply(lambda x:len(x))
frequent_items
```

```python
# tìm mối liên quan giữa các mặt hàng, hay được bán cùng nhau
from mlxtend.frequent_patterns import association_rules
```

```python
final = association_rules(frequent_items, metric='lift',
                 min_threshold=1).sort_values('lift',ascending=False).reset_index(drop=True)
final
```

### Định nghĩa các thông số
1. lift: sự tương quan giữa 2 item, nếu >1 là có sự tương quan, tức là nó thường được bán chung hơn là bán riêng
2. support: tần suất xuất hiện của item hoặc 1 cặp item (%)
3. confidence: giải thích sự xuất hiện của antecedent (cái xảy ra trước) và consequent (cái xảy ra sau). Tính bằng trong tổng số những txn mua B, thì có bn cái mua cả A và B :))
Nếu support của A > C => nên tính lại số confidence từ C qua A, vì sẽ cao hơn và nên suggest là ng dùng thường sẽ mua A nếu trước đó đã mua C
